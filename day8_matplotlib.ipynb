```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,10)
y = x*x
```

```python
print('X: ',x)
print('Y: ',y)
plt.plot(x,y)
```

```python
x = ['a','b','c','d']
y = [10,20,30,40]
print('X : ',x)
print('Y : ',y)
plt.title('Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y)
```

```python

x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = ['a','b','c','d','e','f']
y = [x for x in range(1,7)]
plt.bar(x,y)
```

```python
x = ['a','b','c','d','e','f']
y = [x for x in range(1,7)]
plt.barh(x,y)
```

```python
xpoints = np.array([1, 8,4,564,54,21,21,54,54,5])
ypoints = np.array([3, 10,41,541,545,12,1,21,21,69])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
categories = np.array(['A','B','C','D'])
x = np.array([10,20,30,40])
y = np.array([5,15,25,35])
z = np.array([2,22,23,25])
bar_width = 0.50
plt.bar(categories,x,bar_width,label='Dataset1')
plt.bar(categories,y,bar_width-0.10,label='Dataset2')
plt.bar(categories,z,bar_width-0.20,label='Dataset3')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Demo graph')
plt.legend()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,autopct='%1.1f%%')
plt.legend()
plt.show() 
```
